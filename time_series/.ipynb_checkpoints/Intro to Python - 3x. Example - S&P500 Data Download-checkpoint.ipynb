# Plot More Graphs

```python
from datetime import datetime

SnP500 = web.get_data_yahoo('^GSPC', '1/1/2000', '14/2/2014')
# Note: Don't bother with the DJI. No one seems to be able to provide the feed.
STI = web.get_data_yahoo('^STI', '1/1/2000', '14/2/2014')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

SnP500['Adj Close'].plot(ax=ax)

crisis_data = [
(datetime(2007, 10, 11), 'Peak of bull market'),
(datetime(2008, 3, 12), 'Bear Stearns Fails'),
(datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, SnP500['Adj Close'].asof(date) + 100),
    xytext=(date, SnP500['Adj Close'].asof(date) + 400),
    arrowprops=dict(facecolor='black'),
    horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 2000])
ax.set_title('S&P500: Important dates in 2008-2009 financial crisis')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

STI['Adj Close'].plot(ax=ax)

crisis_data = [
(datetime(2007, 10, 11), 'Peak of bull market'),
(datetime(2008, 3, 12), 'Bear Stearns Fails'),
(datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, STI['Adj Close'].asof(date) + 100*2),
    xytext=(date, STI['Adj Close'].asof(date) + 400*2),
    arrowprops=dict(facecolor='black'),
    horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([1000, 5000])
ax.set_title('STI: Important dates in 2008-2009 financial crisis')
```
